### Detect Edges

```python
# Calculate median intensity
median_intensity= np.median(image_grey)



#Set threshold to be one standard deviation above and below median intensity
lower_threshold = int(max(0,(1.0-0.33) * median_intensity))

upper_threshold = int(min(255,(1.0+0.33)* median_intensity))

#Apply canny edge detector

image_canny = cv2.Canny(image_normal, 40, 1)
```

```python
plt.imshow(image_canny, cmap='gray'),plt.axis("off")
plt.show()
```

#### Enhance Contrast (COLOR)

```python
image_blood_yuv= cv2.cvtColor(image_blood2, cv2.COLOR_BGR2YUV)
#if you read in using .imread with a second parameter "0", theres no need to use cvtColor (it will break/error)
# second parameter '0'= greyscale
```

#### Enchance image

```python
image_blood_yuv[:,:,0]= cv2.equalizeHist(image_blood_yuv[:,:,0])
```

#### Convert to RBG

```python
image_rgb = cv2.cvtColor(image_blood_yuv, cv2.COLOR_YUV2RGB)
```
